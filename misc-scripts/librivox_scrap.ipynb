```python
import requests
from bs4 import BeautifulSoup
import json
import os
import random

headers = {
    'Access-Control-Allow-Origin': '*',
    'Access-Control-Allow-Methods': 'GET',
    'Access-Control-Allow-Headers': 'Content-Type',
    'Access-Control-Max-Age': '3600',
    'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'
    }

tags = ['Suggestions', 'Recently opened', 'Popular this month', 'Latest', 'All time favorite']
urls = []
with open('librivox-top50.txt', 'r') as f:
    urls = f.readlines()

books = []
for i, url in enumerate(urls):
    bk = {}
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    bk['front_cover'] = soup.find("a", {"class": "download-cover"})['href']
    bk['synopsis'] = soup.find("p", {"class": "description"}).string
    ps = soup.find_all("p", {"class": "book-page-genre"})
    ps[0].span.decompose()
    bk['category'] = ps[0].text
    ps[1].span.decompose()
    bk['language'] = ps[1].text
    bk['title'] = soup.find("div", {"class": "content-wrap clearfix"}).h1.text
    author = soup.find("p", {"class": "book-page-author"}).a
    if author.span:
        author.span.decompose()
    bk['author'] = author.text
    bk['year'] = 2020
    bk['patent'] = 'PD'
    bk['sample'] = None
    bk['rating'] = 0
    bk['price'] = 0
    bk['tagline'] = tags[random.randint(0, len(tags)-1)]
    trs = soup.find("table", {"class": "chapter-download"}).tbody.find_all('tr')
    chapters = []
    for tr in trs:
        tds = tr.find_all('td')
        bc = {}
        bc['audio_file'] =  tds[1].a['href']
        bc['chapter'] = tds[1].a.text
        chapters.append(bc)
    bk['book_chapter'] = chapters
    books.append(bk)
    print("Completed parsing {}".format(url))

dest = 'books.json'
if os.path.exists(dest):
    os.remove(dest)

print("Parsed {} books.".format(len(books)))
with open(dest, 'w') as f:
  json.dump(books, f, indent=4, sort_keys=True)     

print("Completed parsing all {} urls. Saved to {}".format(len(urls), dest))
```
